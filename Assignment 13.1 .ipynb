```python
#define the global function 
pysqldf = lambda q: sqldf(q, globals())
```

```python
#1. Select 10 records from the adult sqladb
pysqldf("SELECT * FROM adult LIMIT 10;")
```

```python
# 2. Show me the average hours per week of all men who are working in private sector
q = """SELECT
            AVG(HoursPerWeek) as AvgHrs
           FROM
              adult
           WHERE 
              WorkClass = 'Private'
              AND
              Sex = 'Male'
           GROUP BY
              Workclass,
              Sex  """
```

```python
pysqldf(q)
```

```python
# 3. Show me the frequency table for education, occupation and relationship, separately

#Frequency table for Relationship
q = """SELECT
            Relationship , Count(*) as Freq
           FROM
              adult
            GROUP BY
            Relationship """
pysqldf(q)
```

```python
#Frequency table for Occupation
q = """SELECT
            Occupation , Count(*) as Freq
           FROM
              adult
            GROUP BY
              Occupation """
pysqldf(q)
```

```python
#Frequency table for Education
q = """SELECT
            Education , Count(*) as Freq
           FROM
              adult
            GROUP BY
              Education   """
pysqldf(q)
```

```python
# 4. Are there any people who are married, working in private sector and having a masters degree

adult["MaritalStatus"].value_counts()
```

```python
q = """SELECT
            count(*)
           FROM
              adult
           WHERE
            MaritalStatus NOT IN ('Never-married', 'Separated', 'Divorced', 'Widowed')
            AND 
            WorkClass ='Private'
            AND
            Education = 'Masters' 
           """
pysqldf(q)
```

```python
# 5. What is the average, minimum and maximum age group for people working in different sectors
q = """SELECT
            WorkClass, AVG(Age) as AvgAge, MIN(Age) as MinAge, MAX(Age) as MaxAge
           FROM
              adult
           GROUP BY WorkClass
           
           """
pysqldf(q)
```

```python
# 6. Calculate age distribution by country
q = """SELECT
            NativeCountry , Age
           FROM
              adult
           GROUP BY NativeCountry,Age
           
           """
pysqldf(q)
```

```python
# 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'
q = """SELECT
            ((CapitalGain)-(CapitalLoss)) as 'Net-Capital-Gain'
           FROM
              adult
           
           """
pysqldf(q)
```
